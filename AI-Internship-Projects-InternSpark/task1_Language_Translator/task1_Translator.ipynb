```python
from deep_translator import GoogleTranslator

print("AI Language Translator")
print("Language Codes: hi=Hindi, fr=French, es=Spanish, de=German")

while True:
    text = input("\nEnter text to translate: ")
    target = input("Enter target language code: ")

    translated = GoogleTranslator(source='auto', target=target).translate(text)

    print("Translated Text:", translated)

    choice = input("Do you want to translate another text? (yes/no): ")

    if choice.lower() != "yes":
        print("Translator Closed.")
        break
```

```python
from deep_translator import GoogleTranslator

print("AI Language Translator")

print("\nLanguage Codes:")
print("en = English")
print("hi = Hindi")
print("fr = French")
print("es = Spanish")
print("de = German")

while True:
    source = input("\nEnter source language code: ")
    target = input("Enter target language code: ")
    text = input("Enter text to translate: ")

    translated = GoogleTranslator(source=source, target=target).translate(text)

    print("Translated Text:", translated)

    choice = input("\nTranslate another text? (yes/no): ")

    if choice.lower() != "yes":
        print("Translator Closed.")
        break
```
