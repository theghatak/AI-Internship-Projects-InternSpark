```python
import random

print("AI FAQ Chatbot")
print("Type 'exit' to stop the chatbot.\n")

# FAQ dataset
faq = {
    "hello": ["Hello! How can I help you?", "Hi there!", "Hello! Nice to meet you."],
    "how are you": ["I am doing well!", "I'm fine, thank you!", "I'm just a bot but I'm great!"],
    "what is your name": ["I am an AI chatbot.", "You can call me AI Bot."],
    "what can you do": ["I can answer simple questions.", "I can help with FAQs."],
    "who created you": ["I was created as part of an AI internship project."],
    "bye": ["Goodbye!", "See you later!", "Have a great day!"]
}

# Chat loop
while True:
    user_input = input("You: ").lower()

    if user_input == "exit":
        print("Bot: Goodbye!")
        break

    elif user_input in faq:
        print("Bot:", random.choice(faq[user_input]))

    else:
        print("Bot: Sorry, I don't understand that question.")
```
